```python
def handle_missing_values(df):
    df = df.replace("?", np.nan)

    for col in df.columns:
        if df[col].dtype == 'object':
            df[col] = df[col].fillna(df[col].mode()[0])
        else:
            df[col] = df[col].fillna(df[col].median())

    return df

```

```python
def encode_categorical_columns(df):
    le = LabelEncoder()
    for col in df.select_dtypes(include=['object']).columns:
        df[col] = le.fit_transform(df[col])
    return df
```

```python
def handle_outliers(df):
    numeric_cols = df.select_dtypes(include=np.number).columns

    for col in numeric_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        df[col] = np.clip(df[col], lower, upper)

    return df
```

```python
def min_max_scaling(df):
    scaler = MinMaxScaler()
    numeric_cols = df.select_dtypes(include=np.number).columns
    df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
    return df
```

```python
def standard_scaling(df):
    scaler = StandardScaler()
    numeric_cols = df.select_dtypes(include=np.number).columns
    df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
    return df
```

```python
diabetes_cleaned =df.copy()

diabetes_cleaned = handle_missing_values(diabetes_cleaned)
diabetes_cleaned = handle_outliers(diabetes_cleaned)

# Scaling
diabetes_minmax = min_max_scaling(diabetes_cleaned.copy())
diabetes_standard = standard_scaling(diabetes_cleaned.copy())

print("Diabetes preprocessing completed")
```

```python
adult_cleaned =df1.copy()

adult_cleaned = handle_missing_values(adult_cleaned)
adult_cleaned = encode_categorical_columns(adult_cleaned)
adult_cleaned = handle_outliers(adult_cleaned)

# Scaling
adult_minmax = min_max_scaling(adult_cleaned.copy())
adult_standard = standard_scaling(adult_cleaned.copy())

print("Adult Income preprocessing completed")
```
